# EDA (exploratory data analysis)

```python
import pandas as pd
df=pd.read_csv('tested.csv')
df.head()
```

# 1 categorical data

# a . count plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=df)
plt.show()
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```
